```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
import numpy as np
import matplotlib.pyplot as plt

# Kuantum devresini tanımla (1 qubit)
qc = QuantumCircuit(1)

# Döndürme açısı 
rotation_angle = 2 * np.pi / 3  # 120 derece dönüş


# Grafik için ekseni çizme fonksiyonu
def draw_axes():
    plt.axhline(0, color='gray', linestyle='--')
    plt.axvline(0, color='gray', linestyle='--')

# Birim çember çizme fonksiyonu
def draw_unit_circle():
    circle = plt.Circle((0, 0), 1, color='black', fill=False)
    plt.gca().add_patch(circle)

# Kuantum durumlarını çizme fonksiyonu
def draw_quantum_state(state, iteration):
    real_part = np.real(state[0])  # Gerçek (Re) kısmı
    imag_part = np.imag(state[0])  # Sanal (Im) kısmı

    # Grafiği çizme
    plt.figure(figsize=(5, 5))
    draw_axes()
    draw_unit_circle()
    
    # Durum vektörünü çiz
    plt.scatter(real_part, imag_part, color='red', label=f'Iteration {iteration}')
    plt.xlim(-1.1, 1.1)
    plt.ylim(-1.1, 1.1)
    plt.xlabel("Re(ψ)")
    plt.ylabel("Im(ψ)")
    plt.title(f"Quantum State after {iteration} Iterations")
    plt.legend()
    plt.grid()
    plt.show()

# 7 iterasyon boyunca Ry(4π) uygula
for i in range(7):
    qc.ry(rotation_angle, 0)  # Ry kapısını uygula

    # Durum vektörünü hesapla
    statevector = Statevector.from_instruction(qc)
    state = statevector.data  # Kompleks sayı içeren vektör

    # Sonucu yazdır
    print(f"Iteration {i+1}: Quantum State = {state}")

    # Kuantum durumunu çiz
    draw_quantum_state(state, i+1)
```
